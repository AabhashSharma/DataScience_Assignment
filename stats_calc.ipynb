### Compute the mean, median, and mode of X

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from statistics import mode

X = np.array([2,4,4,5,6,6,6,7,8,10])
Y = np.array([1,3,4,4,5,7,8,9,10,12])

mean_X = np.mean(X)
median_X = np.median(X)
mode_X = mode(X)

print("Mean of X:", mean_X)
print("Median of X:", median_X)
print("Mode of X:", mode_X)

```

### Compute the variance and standard deviation of X

```python
variance_X = np.var(X, ddof=1)   
std_X = np.std(X, ddof=1)       

print("Sample Variance of X:", variance_X)
print("Sample Standard Deviation of X:", std_X)
```
